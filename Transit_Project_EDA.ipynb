```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("drive/MyDrive/CTA_-_Ridership_-_Bus_Routes_-_Monthly_Day-Type_Averages___Totals.csv")
df['Month_Beginning'] = pd.to_datetime(df['Month_Beginning'], format='%m/%d/%Y')
```

```python
totals = df.groupby(['Month_Beginning']).mean()["MonthTotal"]
```

```python
totals.plot()
```

```python
totals.groupby(totals.index.year).mean().plot()
```

```python
covidless_totals = totals.loc['2001-1-1':'2019-11-1']
covidless_totals.groupby(covidless_totals.index.month).mean().plot()
```
